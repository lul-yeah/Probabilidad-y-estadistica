### distribucion normal

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Datos
media = 70
desv_std = 4
total_estudiantes = 500
limite_inferior = 68
limite_superior = 72

# Probabilidades
prob_inferior = norm.cdf(limite_inferior, media, desv_std)
prob_superior = norm.cdf(limite_superior, media, desv_std)
prob_rango = prob_superior - prob_inferior
estudiantes_rango = total_estudiantes * prob_rango

# Resultados
print(f"Probabilidad de que un estudiante pese entre 68kg y 72kg: {prob_rango:.4f}")
print(f"Número de estudiantes que pesan entre 68kg y 72kg: {estudiantes_rango:.0f} de 500")

# Gráfica
x = np.linspace(media - 4*desv_std, media + 4*desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.figure(figsize=(8, 5))
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= limite_inferior) & (x <= limite_superior), color='lightgreen', alpha=0.6)
plt.title(f'Distribución Normal del Peso de Estudiantes ITD\n'
          f'$\mu$ = {media}, $\sigma$ = {desv_std}\n'
          f'Probabilidad entre 68kg y 72kg: {prob_rango:.2%} → Estudiantes ≈ {estudiantes_rango:.0f}')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

### binomial

Desarrollo de Problema 2 en formato Binomial
El problema 2 es de tipo Binomial porque:

Involucra un número fijo de pruebas (n = 1000).

Cada prueba tiene dos resultados: motor defectuoso o no defectuoso.

La probabilidad de defecto es constante (p = 0.001).

Esto cumple con las condiciones de una distribución binomial.

```python
# PROBLEMA 2 - Distribución Binomial

from scipy.stats import binom
import matplotlib.pyplot as plt

# Parámetros
n = 1000        # Total de motores
p = 0.001       # Probabilidad de que un motor sea defectuoso
q = 1 - p       # Probabilidad de que un motor NO sea defectuoso

# a) Probabilidad de que ninguno sea defectuoso
x0 = 0
prob_ninguno = binom.pmf(x0, n, p)
print(f"a) Probabilidad de que ninguno sea defectuoso: {prob_ninguno:.5f}")

# b) Probabilidad de que tres sean defectuosos
x1 = 3
prob_tres = binom.pmf(x1, n, p)
print(f"b) Probabilidad de que tres sean defectuosos: {prob_tres:.5f}")

# c) Probabilidad de que cuatro sean defectuosos
x2 = 4
prob_cuatro = binom.pmf(x2, n, p)
print(f"c) Probabilidad de que cuatro sean defectuosos: {prob_cuatro:.5f}")

# Gráfica de la distribución binomial para x de 0 a 10 defectuosos
x = range(11)
y = binom.pmf(x, n, p)

plt.figure(figsize=(8, 5))
plt.bar(x, y, color='skyblue', edgecolor='black')
plt.title("Distribución Binomial - Motores defectuosos en un lote de 1000")
plt.xlabel("Número de motores defectuosos")
plt.ylabel("Probabilidad")# Unidad 4

## Tarea


### facilitador :dr jose gabriel rodriguez rivas

### alumno : Josue hernadez castro
plt.xticks(x)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

## Poison
Problema 1 — Distribución de Poisson

Este problema es de tipo Poisson porque:

Trata sobre la cantidad de eventos en un intervalo fijo de tiempo (personas ingresadas en un día).

Los eventos son raros e independientes entre sí.

Se conoce la tasa promedio de ocurrencia (λ = 5).

Por lo tanto, se modela con una distribución de Poisson con λ = 5.



```python
from scipy.stats import poisson
import matplotlib.pyplot as plt
import numpy as np

# Parámetros
media = 5  # λ = 5 personas ingresadas por día

print("PROBLEMA 1")
print("Distribución de Poisson: λ = 5")

# a) Probabilidad de que ingresen exactamente 2 personas
x = 2
p_x_2 = poisson.pmf(x, media)
print(f"\na) Probabilidad de que ingresen exactamente 2 personas: {p_x_2:.5f}")

# b) ¿Es probable que x exceda de 10?
# Se calcula como 1 - P(x ≤ 10)
p_mayor_10 = 1 - poisson.cdf(10, media)
print(f"\nb) Probabilidad de que ingresen más de 10 personas: {p_mayor_10:.5f}")

if p_mayor_10 > 0.05:
    print("→ Es probable que ingresen más de 10 personas.")
else:
    print("→ No es muy probable que ingresen más de 10 personas.")

# Gráfica de la distribución Poisson para λ = 5
valores_x = np.arange(0, 16)
valores_y = poisson.pmf(valores_x, media)

plt.figure(figsize=(8, 5))
plt.bar(valores_x, valores_y, color='mediumseagreen', edgecolor='black')
plt.title("Distribución de Poisson (λ = 5)\nIngresos diarios a UCI")
plt.xlabel("Número de ingresos")
plt.ylabel("Probabilidad")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.xticks(valores_x)
plt.show()
```
