```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from scipy import stats

# Configuración de estilo para los gráficos
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

def analizar_grupo(nombre_archivo, grupo_num):
    # Cargar los datos
    df = pd.read_csv(nombre_archivo)

    print(f"\n{'='*50}")
    print(f"Análisis para Grupo {grupo_num}")
    print(f"{'='*50}\n")

    # 1. Exploración inicial de los datos
    print("1. Descripción estadística de los datos:")
    print(df.describe())

    # 2. Histograma de calificaciones
    plt.figure(figsize=(10, 6))
    sns.histplot(df['Calificacion'], bins=15, kde=True)
    plt.title(f'Distribución de Calificaciones - Grupo {grupo_num}')
    plt.xlabel('Calificación')
    plt.ylabel('Frecuencia')
    plt.show()

    # 3. Modelo de regresión lineal
    X = df['HorasEstudio'].values.reshape(-1, 1)
    y = df['Calificacion'].values

    modelo = LinearRegression()
    modelo.fit(X, y)

    pendiente = modelo.coef_[0]
    intercepto = modelo.intercept_
    r_cuadrado = modelo.score(X, y)

    print("\n2. Modelo de Regresión Lineal:")
    print(f"Ecuación: Calificación = {pendiente:.4f} * HorasEstudio + {intercepto:.4f}")
    print(f"Coeficiente de determinación (R²): {r_cuadrado:.4f}")

    # 4. Coeficiente de correlación de Pearson
    r_pearson, p_valor = stats.pearsonr(df['HorasEstudio'], df['Calificacion'])
    print(f"\n3. Coeficiente de correlación de Pearson: {r_pearson:.4f}")
    print(f"Valor p: {p_valor:.4f}")

    # 5. Visualización de resultados
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='HorasEstudio', y='Calificacion', data=df, label='Datos observados')

    # Línea de regresión
    x_values = np.linspace(df['HorasEstudio'].min(), df['HorasEstudio'].max(), 100)
    y_values = pendiente * x_values + intercepto
    plt.plot(x_values, y_values, color='red', label=f'Regresión: y = {pendiente:.2f}x + {intercepto:.2f}')

    plt.title(f'Relación entre Horas de Estudio y Calificación - Grupo {grupo_num}')
    plt.xlabel('Horas de Estudio')
    plt.ylabel('Calificación')
    plt.legend()
    plt.grid(True)
    plt.show()

    return {
        'pendiente': pendiente,
        'intercepto': intercepto,
        'r_cuadrado': r_cuadrado,
        'r_pearson': r_pearson,
        'p_valor': p_valor
    }

# Analizar los tres grupos
resultados_g1 = analizar_grupo('grupo1.csv', 1)
resultados_g2 = analizar_grupo('grupo2.csv', 2)
resultados_g3 = analizar_grupo('grupo3.csv', 3)



```

Ecuación:

Según el modelo, por cada hora adicional de estudio, la calificación solo sube 0.42 puntos, lo cual es un cambio muy pequeño. Lo que más me llamó la atención fue el valor inicial (77.11), lo que indica que incluso sin estudiar, los estudiantes de este grupo ya tienen buenas calificaciones. Me hace pensar que el estudio no está influyendo tanto como en los otros casos.

2. Interpretación del coeficiente de correlación
El valor de correlación de Pearson es 0.30, lo que indica una relación positiva, pero bastante débil. Es decir, sí hay un pequeño patrón donde estudiar un poco más mejora la calificación, pero no es muy fuerte ni muy claro. Esto podría deberse a muchos factores fuera del estudio.

3. Significancia estadística del modelo
El valor p es 0.0429, que sigue siendo menor que 0.05, así que el modelo es estadísticamente significativo. Sin embargo, al ser tan cercano al límite, es una significancia “justa”, no tan contundente como en otros grupos.

4. Confiabilidad del modelo (R²)
El R² es 0.09, lo que quiere decir que solo el 9% de las variaciones en las calificaciones se explican por las horas de estudio. El resto (91%) se debe a otros factores. Esto me pareció importante porque aunque el modelo es estadísticamente válido, en la práctica no es muy útil para predecir calificaciones.
