```python
from statistics import*
```

```python
#lista de calificaciones del grupo a de probabilidad y estadistica
grupoa=[70,70,65,90,100,95,45,80,73,20,100,96]
print(grupoa)
```

```python
# sacar la media de las calificaciones con una formula
promedio=sum(grupoa)/len(grupoa)
print(promedio)
```

```python
# utilizar la funcion mean que esta en la libreria para sacar
# la mediana
median_grouped=median(grupoa)
print(median_grouped)
```

```python
# utilizar la funcion mean que esta en la libreria para sacar
# la moda
mode_grouped=mode(grupoa)
print(mode_grouped)
```

```python
multimodass=multimode(grupoa)
print(multimodass)
#
```

```python

import random
fabricas1=[]
for i in range(100):
  numero_aleatorio=round(random.uniform(50,100),2)
  fabricas1.append(numero_aleatorio)
print(fabricas1)
```

```python
import random
fabricas2=[]
for i in range(100):
  numero_aleatorio=round(random.uniform(50,100),2)
  fabricas2.append(numero_aleatorio)
print(fabricas2)
```

```python
#libreria de python para graficar valores
import matplotlib.pyplot as plt
```

```python
plt.hist(fabricas1)
plt.show("produccion diaria de la fabrica 1")
```

```python
plt.hist(fabricas2)
plt.show("produccion diaria de la fabrica 2")
```

```python
mediana1=median(fabricas1)
print(mediana1)
media1=mean(fabricas1)
print(media1)
moda1=mode(fabricas1)
print(moda1)
```

```python
mediana2=median(fabricas2)
print(mediana1)
media2=mean(fabricas2)
print(media1)
moda2=mode(fabricas2)
print(moda2)
```

```python
#generar 2 subplot de 1 renglon por 2 columnas para desplegar los dos histgramas # uno al lado de otro
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
ax1.hist(fabricas1)
ax2.hist(fabricas2)
plt.show()
plt.show()
plt.show()
```

```python
plt.hist(fabricas1, bins=10, alpha=0.2, color="b" , edgecolor= "black" )
plt.axvline(media1, color="b", linestyle="dashed", linewidth=2, label="media1")
plt.axvline(mediana1, color="r", linestyle="dashed", linewidth=2, label="mediana1")
plt.axvline(moda1, color="g", linestyle="dashed", linewidth=2, label="moda1")
plt.xlabel("valor")
plt.ylabel("frecuencia")
plt.title("produccion diaria de la fabrica 1")
plt.legend()
plt.show()
```

```python
plt.hist(fabricas2, bins=10, alpha=0.2, color="b" , edgecolor= "black" )
plt.axvline(media2, color="b", linestyle="dashed", linewidth=2, label="media1")
plt.axvline(mediana2, color="r", linestyle="dashed", linewidth=2, label="mediana1")
plt.axvline(moda2, color="g", linestyle="dashed", linewidth=2, label="moda1")
plt.xlabel("valor")
plt.ylabel("frecuencia")
plt.title("produccion diaria de la fabrica 2")
plt.legend()
plt.show()
```
