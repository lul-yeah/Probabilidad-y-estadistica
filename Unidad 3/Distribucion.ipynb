```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
n = 3
p = 0.5
x_values = [0, 1, 2, 3]
pmf_values = binom.pmf(x_values, n, p)
for x, pmf in zip(x_values, pmf_values):
    print(f"P(X = {x}) = {pmf:.3f}")
```

```python
n = 3
p = 0.5
x = np.arange(0, n + 1)
pmf = binom.pmf(x, n, p)
plt.bar(x, pmf)
plt.xlabel('Numero de soles')
plt.ylabel('Probabilidad')
plt.title('Distribucion Binomial')
plt.show()
```
